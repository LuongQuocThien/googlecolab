```
#@title **Chống bị ngắt kết nối:**
#@markdown Vui lòng chỉ tick 1 ô để không bị lỗi .
 
import time ,random ,string ,datetime ,urllib.request ,json ,os ,sys
 
Sleep = False #@param {type:'boolean'}
Random_request_and_string= True #@param {type:'boolean'}
Console_code = False #@param {type:'boolean'} 
 
os_random = ["Android" ,"IOS"]
locale_random = ["en_US" ,"vi_VN" ,"es_ES" ," en_GB" ," tr_CY" ," yi_US"]
referrer = "aefc642b-b0b8-4f5d-8b03-61a85a09d967"
 
def get_random_string(length):
    letters = string.ascii_lowercase
    result_str = ''.join(random.choice(letters) for i in range(length))
    print(result_str)
 
def genString(stringLength):
    try:
        letters = string.ascii_letters + string.digits
        return ''.join(random.choice(letters) for i in range(stringLength))
    except Exception as error:
        print(error)            
        
def digitString(stringLength):
    try:
        digit = string.digits
        return ''.join((random.choice(digit) for i in range(stringLength)))    
    except Exception as error:
        print(error)    
url = f'https://api.cloudflareclient.com/v0a{digitString(3)}/reg'
 
def run():
    os_done = random.choice(os_random)
    locale_done = random.choice(locale_random)
    try:
        install_id = genString(22)
        body = {"key": "{}=".format(genString(43)),
                "install_id": install_id,
                "fcm_token": "{}:APA91b{}".format(install_id, genString(134)),
                "referrer": referrer,
                "warp_enabled": True,
                "tos": datetime.datetime.now().isoformat()[:-3] + "+05:00",
                "type": os_done,
                "locale": locale_done}
        data = json.dumps(body).encode('utf8')
        headers = {'Content-Type': 'application/json; charset=UTF-8',
                    'Host': 'api.cloudflareclient.com',
                    'Connection': 'Keep-Alive',
                    'Accept-Encoding': 'gzip',
                    'User-Agent': 'okhttp/3.12.1'
                    }
        req         = urllib.request.Request(url, data, headers)
        response    = urllib.request.urlopen(req)
        status_code = response.getcode()    
        return status_code
    except Exception as error:
        print (error)
        
if Sleep:
    count = 10000001
    ! sleep 7890
 
if Random_request_and_string:
    count = 0
 
if Console_code:
  count = 10000001
  print ('function ClickConnect(){')
  print ('  colab.config')
  print ('  console.log("Im Working");')
  print ('  document.querySelector("#top-toolbar > colab-connect-button").shadowRoot.querySelector("#connect").click();')
  print ('  document.querySelector("colab-connect-button").click()')
  print ('  document.querySelector("paper-button#comments").click()')
  print ('  document.querySelector("paper-icon-button").click()')
  print ('  console.log("Im Working");')
  print ('};')
  print ('setInterval(ClickConnect, 60000)')
  print ("\n")
  print ("Nhấn F12 và dán vào console trình duyệt của bạn")
 
while (count < 10000000):
    number = random.randint(1,15)
    print (get_random_string(number))
    print ("\n")
    time.sleep (number)
    result = run()
    ! curl ipecho.net/plain
    time.sleep (5)
    ! curl bot.whatismyipaddress.com
    time.sleep (5)
    ! curl icanhazip.com
    time.sleep (5)
    ! top -bn 1 2>&1 1> /tmp/topN
    time.sleep (5)
    with open('/content/log.txt', 'a') as script:
        script.write(str(number))
    time.sleep (number)
    ! curl --upload-file log.txt https://transfer.sh/log.txt 
    time.sleep (10)
```
